```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import json

df = pd.read_csv('/Users/maxiaoxue/Desktop/data.csv')

date = df['Date'].str.split('-',expand=True)[0]
df['year']= date
d2 = df[df["year"] == '2017']
d2 = d2[d2["Region"] == 'global']
#print(d2)

data1 = d2.groupby(by='Track Name')['Streams'].sum()
sort1 = data1.sort_values(ascending=False).head(10)

print('The top 10 tracks in the global throughout year 2017 with their total stream counts:')
print('------------------------------------------------------------------------------------')
print (sort1)

#List the top 10 tracks in the global throughout year 2017 with their total stream counts.

```

```python
topName = []
group = df.groupby(by='Artist')['Streams'].sum()
sort2 = group.sort_values(ascending=False).head(10)
#List the top 10 artists (or groups) those has the most stream counts for all their tracks

topName = np.array(sort2.keys())
print('The top 10 artists those has the most stream counts for all their tracks:')
print('-------------------------------------------------------------------------')
print(sort2)
#print(topName)

sum = df.groupby([df['Artist'], df['Track Name']])['Streams'].sum()
order = DataFrame({'Stream counts' : sum}).reset_index() 
order = order.sort_values(by = 'Stream counts',ascending=False)
print('\n\nCombined with the stream counts of each of their tracks:')
print('--------------------------------------------------------')
order[order['Artist'].isin(topName)].sort_values(by='Artist',ascending=False)


```

```python
with open('/Users/maxiaoxue/Desktop/countries.json', 'r') as f:
    temp = json.loads(f.read())

specificDate = df['Date'].str.split('-',expand=True)
df['sYear'] = df['Date'].str.split('-',expand=True)[0]
df['sMonth'] = df['Date'].str.split('-',expand=True)[1]
df['sDay'] = df['Date'].str.split('-',expand=True)[2]

specificdf = df[['Track Name','Streams','Date','Region']][(df.sYear  == '2017') & (df.sMonth == '12')]
#print(specificdf)
#print(temp)
NA = []
EU = []
AS = []
SA = []
OC = []

for key,value in temp.items():
    vals = dict(value)
    #print(vals)
    for k,v in vals.items():
        if((k =='continent') & (v == 'NA')):
            NA.append(key.lower())
        elif ((k =='continent') & (v == 'EU')):
            EU.append(key.lower())
        elif ((k =='continent') & (v == 'AS')):
            AS.append(key.lower())
        elif ((k =='continent') & (v == 'SA')):
            SA.append(key.lower())
        elif ((k =='continent') & (v == 'OC')):
            OC.append(key.lower())
            
#print(NA)
#print(EU)
#print(AS)
#print(SA)
#print(OC)
            
#遍历json 找出continent=NA的key
NAgroup = specificdf[specificdf['Region'].isin(NA)]
groupNA = NAgroup.groupby(specificdf['Track Name'])['Streams'].sum()
sortNA = groupNA.sort_values(ascending=False).head(10)

EUgroup = specificdf[specificdf['Region'].isin(EU)]
groupEU = EUgroup.groupby(specificdf['Track Name'])['Streams'].sum()
sortEU = groupEU.sort_values(ascending=False).head(10)

ASgroup = specificdf[specificdf['Region'].isin(AS)]
groupAS = ASgroup.groupby(specificdf['Track Name'])['Streams'].sum()
sortAS = groupAS.sort_values(ascending=False).head(10)

SAgroup = specificdf[specificdf['Region'].isin(SA)]
groupSA = SAgroup.groupby(specificdf['Track Name'])['Streams'].sum()
sortSA = groupSA.sort_values(ascending=False).head(10)

OCgroup = specificdf[specificdf['Region'].isin(OC)]
groupOC = OCgroup.groupby(specificdf['Track Name'])['Streams'].sum()
sortOC = groupOC.sort_values(ascending=False).head(10)
print('The top 10 tracks in December, 2017 for each continent are:')
print('(North America, Europe, Asia, South America, Oceania)')

print('\nNorth America:')
print('---------------------------------------------')
print(sortNA)
print('\n')
print('Europe:')
print('---------------------------------------------')
print(sortEU)
print('\n')
print('Asia:')
print('---------------------------------------------')
print(sortAS)
print('\n')
print('South America:')
print('---------------------------------------------')
print(sortSA)
print('\n')
print('Oceania:')
print('---------------------------------------------')
print(sortOC)
print('\n')
```
